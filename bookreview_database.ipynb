In this exercise we’re going to use the BookReview database.

BookReview is a fictional social cataloging website that allows individuals to search its database of books, and rate them (available to registered users only).

`The BookReview database consists of the following tables:`

   * Books - Provides details about the various books

   * Authors - Provides further details about the author of each book

   * Languages - Provides further details about the language of each book

   * Publishers - Provides further details about the publisher of each book

   * Users - Provides details about BookReview’s registered users

   * Locations - Provides further details about the location of each user

   * Genders - Provides further details about the gender of each user

   * Occupations - Provides further details about the occupation of each user

   * Ratings - Provides details about the ratings each user has made



## 1a. Write a query to display the first_name, last_name, email, birthdate, and gender for all users


```python
q1 ="""
select users.first_name, users.last_name, users.email, users.birthdate, genders.gender
from users
join genders 
on genders.gender_id = users.gender_id
"""
df1 = pd.read_sql(q1,engine)
df1
```

## 1b. Restrict your query to display only users who were born on 2002


```python
q2 ="""
select users.first_name, users.last_name, users.email, users.birthdate, genders.gender
from users
join genders
on genders.gender_id = users.gender_id
where year(birthdate) = 2002
"""
df2 = pd.read_sql(q2,engine)
df2
```

## 1c. Write a query to display the number of users who were born each year. Sort the output by year (ascending)


```python
q3 ="""
select year(birthdate) as year, count(*) as user_num
from users
group by year(birthdate)
order by year asc
"""
df3 = pd.read_sql(q3,engine)
df3
```


## 1d. Repeat your last query, only this time display only years with more than 54 users (who were born on that year)

```python
q4 ="""
select year(birthdate) as year, count(*) as user_num
from users
group by year(birthdate)
having count(*) > 54
"""
df4 =pd.read_sql(q4,engine)
df4
```

## 2c. Write a query to break down the number of users living in each country and city. Sort the output by the number of users (descending)


```python
q7 ="""
select l.country, l.city, count(*) as num_of_user
from users u
join locations l on u.location_id = l.location_id
group by country, city
order by count(*) desc
"""
df7 =pd.read_sql(q7,engine)
df7
```

## 2d. Restrict your query to count only users above the age 30

```python
q8="""
select l.country, l.city, count(*) as num_of_user
from users u
join locations l on u.location_id = l.location_id
where timestampdiff(year,birthdate,current_date()) > 30
group by country, city
order by count(*) desc
"""
df8 =pd.read_sql(q8,engine)
df8
```

## 3b. Write a query to display the average age for each occupation


```python
q10 ="""
select o.occupation_name, avg(timestampdiff(year,birthdate,current_date())) as avg_age                         
from users u
join occupations o
on u.occupation_id = o.occupation_id
group by occupation_name
"""
df10 =pd.read_sql(q10,engine)
df10
```

## 3c. Restrict your query to display only occupations with average age greater than 40

```python
q11 ="""
select o.occupation_name, avg(timestampdiff(year,birthdate,current_date())) as avg_age                         
from users u
join occupations o
on u.occupation_id = o.occupation_id
group by occupation_name
having avg(timestampdiff(year,birthdate,current_date())) > 40
"""
df11 =pd.read_sql(q11,engine)
df11
```

## 4. Write a query to display the average num_pages for each author. Sort the output by the average number of pages (descending)

```python
q12="""
select a.authors, avg(b.num_pages) as avg_pages
from books b
join authors a
on b.author_id = a.author_id
group by authors
order by avg(b.num_pages) desc
"""
df12 =pd.read_sql(q12,engine)
df12
```

## 6. Write a query to display the number of unique publishers by each language, Sort the output by the number of unique publishers (descending)

```python
q14="""
select l.language_name, count(distinct(publisher_id)) as number
from books b
join languages l
on b.language_code = l.language_code
group by language_name
order by count(distinct(publisher_id)) desc
"""
df14 =pd.read_sql(q14,engine)
df14
```

## 8a. Write a query to display the rating_id, rating_date, rating, book title, and user's full name for each rating

## 8b. Restrict your query to display ratings related to the 'Harry Potter' book-series


```python
q18="""
select concat(u.first_name," ",u.last_name) as full_name, r.rating_id, r.rating_date, r.rating, b.title
from ratings r
join users u on u.user_id = r.user_id
join books b on r.book_id = b.book_id
where instr(b.title,'harry potter') > 0
"""
df18 = pd.read_sql(q18,engine)
df18
```

## 8c. How many unique users have rated titles related to the 'Harry Potter' book-series?

```python
q19 ="""
select count(distinct(u.user_id)) as unique_users_number
from users u
join ratings r on u.user_id = r.user_id
join books b on r.book_id = b.book_id
where instr(b.title,'harry potter') > 0
"""
df19 = pd.read_sql(q19,engine)
df19
```

## 9. How many unique users have rated titles related to the 'Harry Potter' book-series. Break down your result by each country

```python
q20 ="""
select l.country, count(distinct(u.user_id)) as unique_users_number
from users u
join ratings r on u.user_id = r.user_id
join books b on r.book_id = b.book_id
join locations l on u.location_id = l.location_id
where instr(b.title,'harry potter') > 0
group by l.country
"""
df20 =pd.read_sql(q20,engine)
df20
```
