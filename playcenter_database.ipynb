## Write a query to display all of the information inside players

```python
q1="""select * from players"""
df1=pd.read_sql(q1,engine)
df1
```

## Write a query to retrieve the last_name, first_name and city for each player

```python
q2="""
select first_name, last_name, city
from players"""
df2=pd.read_sql(q2,engine)
df2
```

## Write a query to retrieve the email_address, country, and street_address for each player

```python
q3="""
select email_address, street_address, country
from players"""
df3=pd.read_sql(q3,engine)
df3
```

## Write a query to display the player_id, total_playing_minutes, and installed_games for each player.

### The installed_games column represents the amount of games each player has purchased/installed

```python
q4="""
select player_id,total_playing_minutes, installed_games
from players"""
df4=pd.read_sql(q4,engine)
df4
```

## Write a query to display the player_id, age_group, and total_playing_minutes for each player

```python
q5="""
select player_id, age_group, total_playing_minutes
from players"""
df5=pd.read_sql(q5,engine)
df5
```

## Write a query to display for each player - the player_id, amount_spent_usd, and a new calculation representing the amount spent in euros, with exchange rate of 0.8.

### Allow us to elaborate, The amount_spent_usd column represents the amount of money spent by each player in USD so if the current exchange rate is 0.8, the calculation would be amount_spent_usd * 0.8.

```python
q6="""
select player_id, amount_spent_usd, (amount_spent_usd * 0.8) as amount_spent_euro
from players"""
df6=pd.read_sql(q6,engine)
df6
```

## Write a query to display the player_id, last_name, first_name, amount_spent_usd, installed_games, and a new calculation representing the average amount spent on each game (amount_spent_usd / installed_games)

```python
q7="""
select player_id, first_name, last_name, amount_spent_usd, (amount_spent_usd/installed_games) as avg_amount_spent
from players"""
df7=pd.read_sql(q7,engine)
df7
```

## Write a query to display for each player - the player_id, email_address, installed_games, uninstalled_games, and a new calculation representing the ratio between installed and uninstalled games (uninstalled_games / installed_games)

```python
q8="""
select player_id,email_address,installed_games,uninstalled_games,(uninstalled_games/installed_games) as installation_ratio
from players"""
df8=pd.read_sql(q8,engine)
df8
```

## Write a query to display the:

### a. player_id

### b. player's first_name concatenated with his/hers last_name

```python
q9="""
select player_id, concat(first_name," ",last_name) as name
from players"""
df9=pd.read_sql(q9,engine)
df9
```

## Write a query to display the:

### a. player_id

### b. email_address

### c. country concatenated with city and street_address

### d. total playing time in hours (total_playing_minutes / 60)

```python
q10="""
select player_id, email_address,concat(street_address,",",city,",",country) as address,(total_playing_minutes/60) as "play_time (hour)"
from players"""
df10=pd.read_sql(q10,engine)
df10
```

## Write a query to display the following string for each player :

### ‘full name preferred language is preferred_language’

### (full name is the result of concatenating first_name and last_name)

### for example:

### 'John Watson preferred language is English'

```python
q11="""
select concat(first_name," ",last_name," preferred language is ",preferred_language) as language_preferred
from players"""
df11=pd.read_sql(q11,engine)
df11
```

## Write a query to display the distinct values inside the preferred_language column

```python
q12="""
select count(distinct preferred_language) as total_unique_value
from players"""
df12=pd.read_sql(q12,engine)
df12
```
